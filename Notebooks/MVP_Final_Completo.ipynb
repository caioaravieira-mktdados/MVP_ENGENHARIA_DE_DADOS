MVP ENGENHARIA DE DADOS –  CAIO AUGUSTO RODRIGUES ALVES VIEIRATema: Análise de Desempenho Estudantil1. Objetivo do ProjetoO objetivo deste MVP é construir um pipeline de dados utilizando as camadas Bronze, Prata e Ouro no Databricks, a partir do dataset público 'Students Performance', obtido pelo Kaggle. Aqui buscaremos entender como as variáveis socioeducacionais influenciam as notas dos estudantes em matemática, leitura e escrita. A ideia é responder às seguintes perguntas:- Existe diferença de desempenho entre gêneros?- A escolaridade dos pais influencia nas notas?- O curso preparatório melhora o desempenho?- Existe alguma diferença de performance entre etnias?- Há correlação entre as notas de matemática, leitura e escrita?2. Coleta dos DadosA coleta foi realizada por upload do arquivo no Databricks. Por ser um dataset público, não houve necessidade de web scraping ou coleta automatizada.3. Modelagem dos DadosA modelagem adotada segue o padrão de Data Lake com três camadas:- Bronze: dados brutos, qualquer alteração da base extraída;- Prata: dados tratados, padronizados e preparados para análise;- Ouro: tabelas analíticas e métricas consolidadas.Para a camada Ouro, foi construído um modelo dimensional simples (Esquema Estrela), contendo:Fato:- Fato_Performance (nota de matemática, leitura, escrita, ID do aluno)Dimensões:- Dim_Genero- Dim_Etnia- Dim_Educacao_PaisCatálogo de Dados:Genero (string) — male, femaleEtnia (string) — group A, B, C, D, EEducacao_dos_pais (string) — níveis educacionaisTipo_Almoco — standard, free/reducedCurso_Preparatorio — none, completedNotas — inteiros de 0 a 1004. Carga dos Dados (ETL no Databricks)A carga foi realizada por meio de notebooks PySpark.5. AnáliseQualidade dos Dados:Sem nulos, sem duplicados, notas entre 0 e 100 → qualidade boa.Soluções:- Gênero: mulheres melhores em leitura/escrita; homens melhores em matemática.- Escolaridade dos pais: quanto maior, maiores as notas.- Curso preparatório: melhora desempenho.- Etnia: grupo E melhor desempenho, grupo A o pior.- Correlação: leitura/escrita = 0,95; escrita/matemática = 0,80.6. ConclusãoPipeline construído com sucesso, camadas organizadas, análises coerentes, objetivos atendidos.7. AutoavaliaçãoPipeline funcionou bem, Databricks exigiu atenção, futuras melhorias incluem automação, dashboards e modelos preditivos.
